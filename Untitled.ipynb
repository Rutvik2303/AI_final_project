```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
from tensorflow import keras

# Load the dataset into a Pandas dataframe
df = pd.read_csv('test.csv')
```

```python
#checking and cleaning/Normalize the data
```

```python
# Check for missing values
print(df.isnull().sum())

# Plot histograms of input features
df.hist(figsize=(12,15))
plt.show()

# Calculate output label distribution
label_counts = df['satisfaction'].value_counts()
print(label_counts)
```

```python
#here we are dropping two column because those are unnecessary like index and customer id
#secondly, i am mapping all string value in column to int like 0 and 1 as well as we are feeling all null with mean
```

```python
df.drop(columns=['id','Unnamed: 0'], inplace=True)
df['Gender'].replace(['Male','Female'],[0,1],inplace=True)
df['Customer Type'].replace(['Loyal Customer','disloyal Customer'],[0,1],inplace=True)
df['Type of Travel'].replace(['Business travel','Personal Travel'],[0,1],inplace=True)
df['Class'].replace(['Eco','Business','Eco Plus'],[0,1,2],inplace=True)
df['satisfaction'].replace(['satisfied','neutral or dissatisfied',' '],[1,0,0],inplace=True)

# Normalize the data using Min-Max scaling

#scaler = MinMaxScaler()
df.fillna(df.mean(), inplace=True)

# fit and transform the data using the scaler
#clum=['Gender','Customer Type','Age','Type of Travel','Class','Flight Distance','Inflight wifi service','Departure/Arrival time convenient','Ease of Online booking','Gate location','Food and drink','Online boarding','Seat comfort','Inflight entertainment','On-board service','Leg room service','Baggage handling','Checkin service','Inflight service','Cleanliness','Departure Delay in Minutes','Arrival Delay in Minutes','satisfaction'] 
#for i in clum:
 #   df_norm[i]= scaler.fit_transform(df[i])
df_norm = (df - df.min()) / (df.max() - df.min())
df_norm = pd.DataFrame(df_norm)
# Re-plot histograms of input features after normalization
df_norm.hist(figsize=(12,15))
plt.show()
print(df_norm.isnull().sum())
```

```python
# Phase 2

#here, we are trying to overfit the entire dataset with three different model
#model 1:- 1,  epochs=100
```

```python
# Load the dataset into a Pandas dataframe
df = df_norm

# Split the dataset into input features (X) and output labels (y)
X = df.drop('satisfaction', axis=1).values
y = df['satisfaction'].values

# Normalize the input features using Min-Max scaling
#X_norm = (X - X.min()) / (X.max() - X.min())
X_norm = X
# Define a logistic regression model with one neuron
model = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[X.shape[1]])
])

# Compile the model with binary cross-entropy loss and Adam optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model for 1000 epochs
history = model.fit(X_norm, y, epochs=100, verbose=0)

# Plot the accuracy and loss curves during training
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['loss'], label='loss')
plt.legend()
plt.show()
```

```python
#model 2:- 2-1,  epochs=500
```

```python
# Define a small neural network with 2 neurons in the input layer and 1 neuron in the output layer
model = keras.Sequential([
    keras.layers.Dense(units=2, input_shape=[X.shape[1]]),
    keras.layers.Dense(units=1)
])

# Compile the model with binary cross-entropy loss and Adam optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model for 1000 epochs
history = model.fit(X_norm, y, epochs=500, verbose=0)

# Plot the accuracy and loss curves during training
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['loss'], label='loss')
plt.legend()
plt.show()
```

```python
#model 3:- 16-8-1,  epochs=1000
```

```python
# Define a larger neural network with 16 neurons in the input layer, 8 neurons in the hidden layer, and 1 neuron in the output layer
model = keras.Sequential([
    keras.layers.Dense(units=16, input_shape=[X.shape[1]]),
    keras.layers.Dense(units=8, activation='relu'),
    keras.layers.Dense(units=1)
])

# Compile the model with binary cross-entropy loss and Adam optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model for 1000 epochs
history = model.fit(X_norm, y, epochs=1000, verbose=0)

# Plot the accuracy and loss curves during training
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['loss'], label='loss')
plt.legend()
plt.show()
```

```python
# Load data
data = df
# Shuffle rows
data = data.sample(frac=1).reset_index(drop=True)

# Split data into training and validation sets
train_data = data.iloc[:int(0.8*len(data))]
val_data = data.iloc[int(0.8*len(data)):]

# Split features and labels
train_features = train_data.drop('satisfaction', axis=1)
train_labels = train_data['satisfaction']
val_features = val_data.drop('satisfaction', axis=1)
val_labels = val_data['satisfaction']

# Convert to numpy arrays
train_features = np.array(train_features)
train_labels = np.array(train_labels)
val_features = np.array(val_features)
val_labels = np.array(val_labels)

# Define the model
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(2, activation='relu', input_shape=(22,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(6, activation='relu', input_shape=(22,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model3 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(22,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])



```

```python
# Compile the model

model= model1 
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model with model checkpointing
checkpoint_path = "model_checkpoints/cp.ckpt"
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path, save_weights_only=True, verbose=1)
history = model.fit(train_features, train_labels, epochs=100, validation_data=(val_features, val_labels), callbacks=[cp_callback])

# Evaluate the model on the training set and validation set
train_loss, train_acc = model.evaluate(train_features, train_labels)
val_loss, val_acc = model.evaluate(val_features, val_labels)

# Make predictions on the validation set
val_pred = model.predict(val_features)
val_pred_labels = np.round(val_pred)

# Calculate other metrics besides accuracy
precision = precision_score(val_labels, val_pred_labels)
recall = recall_score(val_labels, val_pred_labels)
f1 = f1_score(val_labels, val_pred_labels)

# Print results
print(f"Train loss: {train_loss:.4f}, Train accuracy: {train_acc:.4f}")
print(f"Val loss: {val_loss:.4f}, Val accuracy: {val_acc:.4f}")
print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1 score: {f1:.4f}")

# Plot learning curves
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
